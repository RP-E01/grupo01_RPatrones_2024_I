## Preprocesamiento
Búsqueda de nulos o columnas no numéricas

```python
# Chequeo de valores nulos
null_values = data.isnull().sum()

# Chequeo de columnas con tipos de datos no numéricos
non_numeric_columns = data.select_dtypes(exclude=['int64', 'float64']).columns

# Resumen de valores nulos y columnas no numéricas
null_values_summary = null_values[null_values > 0]
non_numeric_summary = data[non_numeric_columns].describe() if len(non_numeric_columns) > 0 else "No hay columnas no numéricas"

null_values_summary, non_numeric_summary
```

Manejo de nulos

```python
# Tratar el valor nulo en 'Geometrical_shape_coefficient' con la mediana
median_value = data['Geometrical_shape_coefficient'].median()
data['Geometrical_shape_coefficient'].fillna(median_value, inplace=True)

# Verificar que el tratamiento haya sido efectivo
data.isnull().sum().sum()
```

## Reducción de dimensionalidad con PCA y kmeans

Se identificó que con 2 componentes se puede explicar el 99.34% de la varianza

```python
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# Aplicar PCA para reducir la dimensionalidad
pca = PCA(n_components=2)  # Reducir a dos dimensiones para visualización y análisis
data_pca = pca.fit_transform(data.drop('cid', axis=1))  # Excluir el identificador único


pca.explained_variance_ratio_.sum()
```

Verificación de la presencia de grupos en base a los 2 componentes identificados:

```python
from sklearn.cluster import KMeans

# Ajustar K-means con 3 clusters basado en los análisis anteriores
kmeans_pca = KMeans(n_clusters=3, random_state=42)
cluster_labels_pca = kmeans_pca.fit_predict(data_pca)

# Gráfico de dispersión de las dos componentes principales
plt.figure(figsize=(8, 6))
scatter = plt.scatter(data_pca[:, 0], data_pca[:, 1], c=cluster_labels_pca, cmap='viridis', edgecolor='k', alpha=0.7)
plt.title('PCA - 2 Principal Components with K-means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

## Determinación de la cantidad óptima de grupos

### Elbow plot

```python

# Rango de número de clusters a probar
range_n_clusters = range(2, 11)

# Lista para guardar las inercias (sumas de las distancias al cuadrado)
inertias = []
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data_pca)
    inertias.append(kmeans.inertia_)

# Elbow plot para visualizar el mejor número de clusters
plt.figure(figsize=(8, 4))
plt.plot(range_n_clusters, inertias, marker='o')
plt.title('Elbow Plot for K-Means')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

### Silhouette score y Davies Bouldin score

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Listas para guardar las métricas de evaluación
silhouette_scores = []
davies_bouldin_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(data_pca)
    silhouette_scores.append(silhouette_score(data_pca, cluster_labels))
    davies_bouldin_scores.append(davies_bouldin_score(data_pca, cluster_labels))

# Gráfico del silhouette score
plt.figure(figsize=(8, 4))
plt.plot(range_n_clusters, silhouette_scores, marker='o', color='blue')
plt.title('Silhouette Scores for K-Means')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Gráfico del Davies-Bouldin score
plt.figure(figsize=(8, 4))
plt.plot(range_n_clusters, davies_bouldin_scores, marker='o', color='green')
plt.title('Davies-Bouldin Scores for K-Means')
plt.xlabel('Number of clusters')
plt.ylabel('Davies-Bouldin Score')
plt.grid(True)
plt.gca().invert_yaxis()  # Invertir el eje y para mostrar que menor es mejor
plt.show()
```

### Gaussian Mixture (maximización de expectativas)

```python
from sklearn.mixture import GaussianMixture

# Lista para guardar los valores BIC
bic_scores = []

for n_components in range_n_clusters:
    gmm = GaussianMixture(n_components=n_components, random_state=42)
    gmm.fit(data_pca)
    bic_scores.append(gmm.bic(data_pca))

# Gráfico del criterio BIC
plt.figure(figsize=(8, 4))
plt.plot(range_n_clusters, bic_scores, marker='o', color='red')
plt.title('BIC Scores for Gaussian Mixture Model')
plt.xlabel('Number of components')
plt.ylabel('BIC Score')
plt.grid(True)
plt.gca().invert_yaxis()  # Invertir el eje y para mostrar que menor es mejor
plt.show()
```

## Empleo de modelos no supervisados para la clasificación

#### Modelos utilizados:
- KMeans
- DBSCAN
- Clustering Jerárquico (ward)

```python
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Ajustar K-means con 4 clusters
kmeans_4 = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans_4.fit_predict(data_pca)

# Ajustar DBSCAN
dbscan = DBSCAN(eps=45, min_samples=8)
dbscan_labels = dbscan.fit_predict(data_pca)

# Ajustar Clustering Jerárquico
hierarchy = linkage(data_pca, method='ward')
hierarchy_labels = fcluster(hierarchy, t=4, criterion='maxclust')

# Crear los subplots
fig, ax = plt.subplots(1, 3, figsize=(18, 6), sharex=True, sharey=True)

# Subplot para K-means
ax[0].scatter(data_pca[:, 0], data_pca[:, 1], c=kmeans_labels, cmap='viridis', edgecolor='k', alpha=0.7)
ax[0].set_title('K-means Clustering (4 Clusters)')
ax[0].set_xlabel('Principal Component 1')
ax[0].set_ylabel('Principal Component 2')

# Subplot para DBSCAN
ax[1].scatter(data_pca[:, 0], data_pca[:, 1], c=dbscan_labels, cmap='viridis', edgecolor='k', alpha=0.7)
ax[1].set_title('DBSCAN Clustering')
ax[1].set_xlabel('Principal Component 1')

# Subplot para Clustering Jerárquico
ax[2].scatter(data_pca[:, 0], data_pca[:, 1], c=hierarchy_labels, cmap='viridis', edgecolor='k', alpha=0.7)
ax[2].set_title('Hierarchical Clustering (Ward)')
ax[2].set_xlabel('Principal Component 1')

plt.tight_layout()
plt.show()
```

## Ploteo de diagrama de cajas para la clasificación en cada modelo

```python
import seaborn as sns

# Crear DataFrame para facilitar los plots
data_for_plots = pd.DataFrame(data_pca, columns=['PC1', 'PC2'])
data_for_plots['KMeans Labels'] = kmeans_labels
data_for_plots['DBSCAN Labels'] = dbscan_labels
data_for_plots['Hierarchical Labels'] = hierarchy_labels

# Crear los boxplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Boxplot para K-means
sns.boxplot(ax=axes[0], x='KMeans Labels', y='PC1', data=data_for_plots, palette='viridis')
axes[0].set_title('K-means Clustering (4 Clusters)')
axes[0].set_xlabel('Cluster Label')
axes[0].set_ylabel('Principal Component 1')

# Boxplot para DBSCAN
sns.boxplot(ax=axes[1], x='DBSCAN Labels', y='PC1', data=data_for_plots, palette='viridis')
axes[1].set_title('DBSCAN Clustering')
axes[1].set_xlabel('Cluster Label')

# Boxplot para Hierarchical Clustering
sns.boxplot(ax=axes[2], x='Hierarchical Labels', y='PC1', data=data_for_plots, palette='viridis')
axes[2].set_title('Hierarchical Clustering (Ward)')
axes[2].set_xlabel('Cluster Label')

plt.tight_layout()
plt.show()
```

## Evaluación final de métricas de los modelos

```python
# Calculando métricas para cada método de clustering
silhouette_kmeans = silhouette_score(data_pca, kmeans_labels)
db_index_kmeans = davies_bouldin_score(data_pca, kmeans_labels)

silhouette_dbscan = silhouette_score(data_pca, dbscan_labels) if len(set(dbscan_labels)) > 1 else 'N/A'
db_index_dbscan = davies_bouldin_score(data_pca, dbscan_labels) if len(set(dbscan_labels)) > 1 else 'N/A'

silhouette_hierarchical = silhouette_score(data_pca, hierarchy_labels)
db_index_hierarchical = davies_bouldin_score(data_pca, hierarchy_labels)

# Imprimir los resultados
print("Métricas de Evaluación para K-means Clustering:")
print(f"Silhouette Score: {silhouette_kmeans:.4f}")
print(f"Davies-Bouldin Index: {db_index_kmeans:.4f}\n")

print("Métricas de Evaluación para DBSCAN Clustering:")
print(f"Silhouette Score: {silhouette_dbscan}")
print(f"Davies-Bouldin Index: {db_index_dbscan}\n")

print("Métricas de Evaluación para Hierarchical Clustering:")
print(f"Silhouette Score: {silhouette_hierarchical:.4f}")
print(f"Davies-Bouldin Index: {db_index_hierarchical:.4f}")
```

## Discusión

K-means y Hierarchical Clustering muestran resultados superiores en términos de Silhouette Score y Davies-Bouldin Index, indicando una mejor separación y cohesión de los clusters comparados con DBSCAN.

DBSCAN tiene scores más bajos, lo que podría reflejar la dificultad de este algoritmo para manejar este conjunto de datos particular sin un ajuste más fino de sus parámetros o posiblemente una necesidad de pre-procesamiento adicional.

La reducción de dimensionalidad fue útil para la simplificación del proceso, debido a que el agrupamiento se puede realizar en base a 2 componentes. Esto sugiere que los datos están muy concentrados o que una gran cantidad de características pueden estar correlacionadas, permitiendo que una gran parte de la variabilidad sea capturada por la primera componente principal.
